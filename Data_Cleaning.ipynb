## Overview
Data cleansing and preprocessing are critical preliminary stages in data analysis and machine learning processes. They convert unrefined, frequently disorganized data into a dependable structure appropriate for modeling and analysis. The process involves identifying and correcting errors, inconsistencies, and inaccuracies in a dataset before analysis or modeling. Typical tasks encompass managing missing data (e.g., by imputation or elimination), eliminating duplicates, rectifying typographical errors or anomalies, and standardizing formats such as dates or text cases. This is to prevent skewed results in analysis, such as biased models from incomplete data.

This outlines a comprehensive data cleaning project utilizing Python's fundamental data science and analytical libraries.

__Key Libraries__

* Pandas 
* Numpy
* Ydata Profiling



## Objective
This project aim to demonstrate data cleaning process of a messy sales dataset by handling missing values, removing duplicates, correcting data types, and generally preparing the dataset for analysis.





## Dataset
The dataset for this project is a synthetic sales dataset generated by AI. It is made up of sales of electronics, computer peripherals, and computer accessories with the following attributes or variables:

1. order_ID: Customer unique identification
2. order_Date: Date a customer placed an order
3. Sales_Rep: Sales representative who handled the order
4. Region: The location of the order
5. Product: The Item being purchased
6. Category: The category of item being purchased
7. Channel: Platform on which the order was made
8. Quantity: The quantity of items being purchased
9. Unit_Price: The price per item
10. Discount_% : The percentage discount on each item
11. Currency: The currency used for the purchase of each item
12. Status: The status of each order





```python
# ===================== Importing the necessary libraries ======================

import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
df= pd.read_excel("C:/Users/Administrator/Desktop/Data_Analyst_Toolkit/messy_sales_data.xlsx")

df.head()
```

```python
# ====================== Profile of the Original Data ======================

df_profile=ProfileReport(df, title="Sales Data Profiling Report", explorative=True)

df_profile
```

####  Insights From the Data Profile Report

The dataset has 20,000 rows and 12 columns covering sales orders from January 2021 to January 2024. The data profile report suggests a widespread quality problems across nearly every field.Some of the quality issues include: 
1. Missing values or empty cells, which is about 13,859.
2. Duplicated rows
3. Unsupported data types
4. Inconsistent categorical casings
5. Typos and casing variants
7. Inconsistent Order_date formats
8. Multiple monetary signs for the Currency column.
9. Trailing whitespaces


```python
# ====================== Summary Data Checks ======================
print("---- Shape of the DataFrame ----")
print(df.shape)         
print('-'*40)

print("---- Data Types of Each Column ----")
print(df.dtypes)        
print('-'*40)

print("---- Missing Values ----")
print(df.isnull().sum())  
print('-'*40)

print("---- Duplicate Rows ----")
print(df.duplicated().sum())
```

```python
# ====================== Removing Duplicates ======================
df=df.drop_duplicates().reset_index(drop=True)
print("Shape after duplicates removed:", df.shape)  
```

```python
# ====================== Converting Numeric Columns ======================

# Removing trailing % from Discount_%
df['Discount_%'] = df['Discount_%'].astype(str).str.replace('%', '', regex=False)

# Converting to numeric where invalid values become NaN
for col in ['Quantity', 'Unit_Price', 'Discount_%']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Getting rid of Negative and  suspicious values
df['Unit_Price'] = df['Unit_Price'].abs()


print(df[['Quantity','Unit_Price','Discount_%']].dtypes)
print(df[['Quantity','Unit_Price','Discount_%']].describe())
```

```python
# ====================== Normalizing Categorical Columns ======================

# Strip leading and trailing whitespace first

str_cols = ['Sales_Rep','Region','Product','Category','Channel','Currency','Status']
for col in str_cols:
    df[col] = df[col].str.strip()

# Title-case for names and readable fields
for col in ['Sales_Rep', 'Region', 'Product', 'Category', 'Channel', 'Status']:
    df[col] = df[col].str.title()

# Using Upper-case for Currency codes
df['Currency'] = df['Currency'].str.upper()
```

```python
# ====================== Inspecting Unique Data Values ======================

for col in ['Region', 'Product', 'Category', 'Channel', 'Status','Currency']:
    print(f"Unique values in {col}: {df[col].unique()}\n")

    
```

* The common issue from these categorical columns points to an issue of inconsistent casing, typos and casing variants. There is also an issue of header leaks where column names appeared as data.

```python
# ======================  Removing Header Leaks ======================
header_leak = (df['Order_ID'] == 'Order_ID') | \
              (df['Region'] == 'Region') | \
              (df['Status'] == 'Status') | \
              (df['Currency'] == 'Currency')
df = df[~header_leak].reset_index(drop=True)
```

```python
# ======================= Fixing Common Typos and Standardizing Categories =======================

#Fixing common typos in Region
#region_map = {'Est':'East','Mw': 'Midwest','MW':'Midwest'}
region_map={'NORTH':'North', 'NOrth':'North', 'North':'North',
              'SOUTH':'South', 'south':'South', 'South':'South',
              'West ':'West', 'West':'West',
              'Est':'East', ' East':'East', 'East':'East',
              'MW':'Midwest','Mw':'Midwest', 'Region':np.nan}
df['Region']= df['Region'].replace(region_map)
print('------- Value Counts for Region -------')
print(df['Region'].value_counts())
print('-'*40)

# Fixing common typos in Product
category_map = {'Electrnics'  : 'Electronics','Periferals'  : 'Peripherals',}
df['Category'] = df['Category'].replace(category_map)
print('------- Value Counts for Category -------')
print(df['Category'].value_counts())
print('-'*40)

# Standardizing Channel values
channel_map = {'In-Store' : 'In Store',}
df['Channel'] = df['Channel'].replace(channel_map)
print('------- Value Counts for Channel -------')
print(df['Channel'].value_counts())

#Standardizing Currency Codes
currency_map = {'US DOLLAR' : 'USD','$': 'USD','CURRENCY': 'USD'}
df['Currency'] = df['Currency'].replace(currency_map)
print('------- Value Counts for Currency -------')
print(df['Currency'].value_counts()) 

# Standardizing Status values
status_map = {'COMPLETED':'Completed', 'completed':'Completed', 'Complete':'Completed',
              'pending':'Pending', 'PENDING':'Pending', 'Cancelled':'Cancelled',
              'cancelled':'Cancelled', 'Refunded':'Refunded', 'Status':np.nan}
df['Status'] = df['Status'].replace(status_map)
print('------- Value Counts for Status -------')
print(df['Status'].value_counts())
```

```python
# ====================== Standardizing Date Format ======================

df['Order_Date'] = pd.to_datetime(df['Order_Date'].astype(str), errors='coerce')
df['Order_Date'] = df['Order_Date'].dt.strftime('%Y-%m-%d')
```

```python
# ====================== Handling Missing Values ======================

# Drop rows missing primary key or date
df = df.dropna(subset=['Order_ID', 'Order_Date', 'Region'])

# Filling string columns
for col in ['Sales_Rep', 'Product', 'Category', 'Channel', 'Status']:
    df[col] = df[col].fillna('N.A')

# Filling null Currency column with most common value (USD)
df['Currency'] = df['Currency'].fillna('USD')

# Filling null numeric columns
df.loc[:, 'Quantity'] = df['Quantity'].fillna(df['Quantity'].median()) 
df.loc[:, 'Unit_Price'] = df['Unit_Price'].fillna(df['Unit_Price'].median())
df.loc[:, 'Discount_%'] = df['Discount_%'].fillna(0)
```

```python
# ====================== Validation Checks ======================

# 1. No duplicates
assert df.duplicated().sum() == 0, 'Duplicates remain'

# 2. No nulls in critical columns
assert df[['Order_ID','Order_Date','Region']].isnull().sum().sum() == 0

# 3. Numeric ranges make sense
assert (df['Quantity'] > 0).all(), 'Negative quantity found'
assert (df['Unit_Price'] > 0).all(), 'Negative unit price found'
assert df['Discount_%'].between(0, 100).all(), 'Discount out of range'

# 4. Categorical values are from expected sets
valid_regions = {'North', 'South', 'East', 'West', 'Midwest'}
assert set(df['Region'].dropna()).issubset(valid_regions), 'Unknown region'

# 5. Date format is consistent
try:
    pd.to_datetime(df['Order_Date'], format='%Y-%m-%d', errors='raise')
except ValueError:
    raise AssertionError('Order_Date has inconsistent format')


# 6. For a consistent currency code, we expect only USD after cleaning
valid_currencies = {'USD'}
assert set(df['Currency'].dropna()).issubset(valid_currencies), 'Unknown currency'

# 7. Status should be from expected set
valid_statuses = { 'Pending', 'Cancelled', 'Refunded', 'Completed', 'N.A' }
assert set(df['Status'].dropna()).issubset(valid_statuses), 'Unknown status'


print('All validation checks passed!')
```
